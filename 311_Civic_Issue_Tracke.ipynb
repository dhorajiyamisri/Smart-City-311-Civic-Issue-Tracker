```python
# 🧹 Data Cleaning

In this section, we will clean the dataset by:

- Checking data types
- Converting date columns
- Handling missing values
- Checking duplicate records
- Preparing the dataset for Exploratory Data Analysis (EDA)
```

```python
# Check Data Types

print("="*60)
print("DATA TYPES")
print("="*60)

display(df.dtypes)
```

```python
# 📅 Convert Date Columns

# The date columns are converted into datetime format to enable time-based analysis.

date_columns = [
    "Created Date",
    "Closed Date",
    "Due Date",
    "Resolution Action Updated Date"
]

for col in date_columns:
    df[col] = pd.to_datetime(df[col], errors="coerce")

print("✅ Date columns converted successfully.")
```

```python
#❓ Missing Value Analysis

# Calculate missing values and their percentages.
missing = pd.DataFrame({
    "Missing Values": df.isnull().sum(),
    "Percentage": round((df.isnull().sum()/len(df))*100,2)
})

missing = missing[missing["Missing Values"]>0]

missing.sort_values("Missing Values",ascending=False)

# Visualize Missing Values

plt.figure(figsize=(10,5))
sns.heatmap(df.isnull(), cbar=False)
plt.title("Missing Values Heatmap")
plt.show()

# Duplicate Check

duplicates = df.duplicated().sum()

print("="*50)
print("Duplicate Records :", duplicates)
print("="*50)

# Remove Duplicates (If Any)

df.drop_duplicates(inplace=True)
print("Dataset Shape :", df.shape)
```

```python
#🔍 Unique Values

# Check the number of unique values in each column.
unique = pd.DataFrame(df.nunique(),columns=["Unique Values"])
unique.sort_values("Unique Values",ascending=False)

# Final Data Information

print("="*60)
print("FINAL DATASET INFORMATION")
print("="*60)

df.info()

# Save Clean Dataset

df.to_csv("311_EDA_Cleaned.csv",index=False)
print("✅ Cleaned Dataset Saved Successfully")
```

```python
# ⚙️ Feature Engineering

Feature Engineering is the process of creating new useful features from existing data.

In this section, we will create new columns that will help us perform deeper analysis, such as:

- Resolution Time
- Resolution Days
- Year
- Month
- Month Number
- Day
- Day Number
- Hour
- Weekend
```

```python
#⏳ Resolution Time

# Calculate the time taken to resolve each complaint.
print("="*60)
df["Resolution Time"] = df["Closed Date"] - df["Created Date"]
print(df[["Resolution Time", "Resolution Days"]])

print("="*60)

#📅 Resolution Days

# Convert resolution time into days for easier analysis.
print("="*60)
df["Resolution Days"] = df["Resolution Time"].dt.days
print(df["Resolution Days"])

print("="*60)
```

```python
# Year
df["Year"] = df["Created Date"].dt.year
print(df[["Created Date", "Year"]])

# Month
df["Month"] = df["Created Date"].dt.month_name()

# Month Number
df["Month Number"] = df["Created Date"].dt.month

# Day Name
df["Day"] = df["Created Date"].dt.day_name()

# Day Number
df["Day Number"] = df["Created Date"].dt.dayofweek

# Hour
df["Hour"] = df["Created Date"].dt.hour

# Weekend
df["Weekend"] = df["Day Number"].apply(
    lambda x: "Weekend" if x >= 5 else "Weekday"
)

print(df[["Created Date", "Month", "Month Number", "Day", "Day Number", "Hour", "Weekend"]])
```

```python
# 📊 Univariate Analysis

Univariate Analysis focuses on analyzing one variable at a time.

In this section, we will explore the distribution of important categorical and numerical variables to understand their patterns and identify key insights.

#🔹 Analysis 1: Complaint Type

# **Objective**
# Identify the most frequently reported complaint types in NYC.
# This analysis helps understand which civic issues are most common and require greater attention from city authorities.
```

```python
print("="*60)
complaint_count = (
    df["Complaint Type"]
    .value_counts()
    .reset_index()
)

complaint_count.columns = ["Complaint Type", "Count"]

display(complaint_count.head(10))
print("="*60)

plt.figure(figsize=(12,6))

sns.barplot(
    data=complaint_count.head(10),
    x="Count",
    y="Complaint Type",
    palette="viridis"
)

plt.title("Top 10 Complaint Types")
plt.xlabel("Number of Complaints")
plt.ylabel("Complaint Type")

plt.show()
print("="*60)

plt.figure(figsize=(8,8))

df["Complaint Type"].value_counts().head(5).plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.ylabel("")
plt.title("Top 5 Complaint Types")

plt.show()
print("="*60)

print("Top 10 Complaint Types")
print("="*60)

display(df["Complaint Type"].value_counts().head(10))

### 💡 Insight

# - A small number of complaint categories contribute to a large portion of total complaints.
# - The city should prioritize these complaint categories to improve public service efficiency.
```

```python
## 🔹 Analysis 2: Agency

**Objective**

Identify which city agencies receive the highest number of complaints.
```

```python
print("="*60)
agency_count = (
    df["Agency"]
    .value_counts()
    .reset_index()
)

agency_count.columns = ["Agency","Count"]

display(agency_count)
print("="*60)

plt.figure(figsize=(10,6))

sns.barplot(
    data=agency_count,
    x="Count",
    y="Agency",
    palette="magma"
)

plt.title("Complaints by Agency")

plt.show()
print("="*60)

#💡 Insight

# Some agencies receive significantly more complaints than others, indicating higher workload and greater public interaction.
```

```python
## 🔹 Analysis 3: Complaint Status

**Objective**

Understand the distribution of complaint status.
```

```python
print("="*60)

status = (
    df["Status"]
    .value_counts()
    .reset_index()
)

status.columns = ["Status","Count"]

display(status)

print("="*60)

plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="Status",
    palette="Set2"
)

plt.title("Complaint Status")

plt.xticks(rotation=30)

plt.show()

print("="*60)

df["Status"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(7,7)
)

plt.ylabel("")
plt.title("Complaint Status Distribution")

plt.show()
print("="*60)


#💡 Insight

# A high percentage of closed complaints indicates good service efficiency, while a large number of open complaints may highlight delays in issue resolution.
```

```python
## 🔹 Analysis 4: Borough

**Objective**

Find which borough generates the highest number of complaints.
```

```python
# 📍 Location & Time Analysis

This section explores where and when complaints are reported.

The objective is to identify:

- Cities with the highest complaints
- Most common address types
- Most common location types
- Monthly complaint trends
- Weekly complaint trends
- Hourly complaint trends
```

```python
## 🔹 Analysis 5: City-wise Complaint Distribution

### Objective

Identify which cities generate the highest number of complaints.
```

```python
print("="*60)
city_count = (
    df["City"]
    .value_counts()
    .head(10)
    .reset_index()
)

city_count.columns = ["City", "Count"]
display(city_count)

print("="*60)
plt.figure(figsize=(12,6))

sns.barplot(
    data=city_count,
    x="Count",
    y="City"
)

plt.title("Top 10 Cities by Complaint Count", fontsize=16)
plt.xlabel("Number of Complaints")
plt.ylabel("City")
plt.show()

### Insight

# The cities with the highest complaint counts may require additional municipal resources and better public service management.
```

```python
## 🔹 Analysis 6: Address Type Distribution

### Objective

Understand how complaints are associated with different address categories.    
```

```python
print("="*60)
address = (
    df["Address Type"]
    .value_counts()
    .reset_index()
)

address.columns = ["Address Type","Count"]
display(address)

print("="*60)
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="Address Type"
)

plt.title("Address Type Distribution")
plt.xticks(rotation=30)
plt.show()

### Insight

# Most complaints are associated with a specific address type, indicating where civic issues are commonly reported.
```

```python
## 🔹 Analysis 7: Location Type Distribution

### Objective

# Identify the most common locations where complaints occur.
print("="*60)
location = (
    df["Location Type"]
    .value_counts()
    .head(10)
    .reset_index()
)

location.columns = ["Location Type","Count"]
display(location)

print("="*60)
plt.figure(figsize=(12,6))

sns.barplot(
    data=location,
    x="Count",
    y="Location Type"
)

plt.title("Top Location Types")
plt.show()

### Insight
# Location type analysis helps identify the environments where civic issues occur most frequently.
```

```python
## 🔹 Analysis 8: Monthly Complaint Trend

### Objective

# Analyze complaint patterns across different months.
print("="*60)
month = (
    df.groupby(["Month Number","Month"])
      .size()
      .reset_index(name="Count")
      .sort_values("Month Number")
)

display(month)    

print("="*60)
plt.figure(figsize=(12,5))
plt.plot(
    month["Month"],
    month["Count"],
    marker="o"
)
plt.xticks(rotation=45)
plt.title("Monthly Complaint Trend")
plt.xlabel("Month")
plt.ylabel("Complaints")
plt.show()    

### Insight

# Monthly trends help identify seasonal patterns in civic complaints.
```

```python
## 🔹 Analysis 9: Weekly Complaint Trend
### Objective
# Identify which day of the week receives the most complaints.

print("="*60)
days = (
    df.groupby(["Day Number","Day"])
      .size()
      .reset_index(name="Count")
      .sort_values("Day Number")
)
display(days)

print("="*60)
plt.figure(figsize=(10,5))

sns.barplot(
    data=days,
    x="Day",
    y="Count"
)

plt.title("Complaints by Day")
plt.show()

### Insight
# Weekly trends help identify operational workload throughout the week.
```

```python
## 🔹 Analysis 10: Hourly Complaint Trend

# Objective
# Identify peak hours when citizens submit complaints.

print("="*60)
hour = (
    df["Hour"]
    .value_counts()
    .sort_index()
)
display(hour)

print("="*60)
plt.figure(figsize=(12,5))
plt.plot(
    hour.index,
    hour.values,
    marker="o"
)

plt.title("Hourly Complaint Trend")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Complaints")
plt.grid(True)
plt.show()

### Insight
# Peak complaint hours can help city authorities schedule support staff more efficiently. 
```

```python
# 🔄 Part 14 - Bivariate Analysis
# Bivariate analysis helps understand the relationship between two variables and provides deeper business insights.

print("="*60)
# Complaint Type vs Borough

top_complaints = df["Complaint Type"].value_counts().head(10).index

cross = pd.crosstab(
    df[df["Complaint Type"].isin(top_complaints)]["Complaint Type"],
    df[df["Complaint Type"].isin(top_complaints)]["Borough"]
)

plt.figure(figsize=(12,6))

sns.heatmap(cross, annot=True, fmt="d", cmap="Blues")

plt.title("Top Complaint Types Across Boroughs")
plt.xlabel("Borough")
plt.ylabel("Complaint Type")
plt.show()

print("="*60)
# Complaint Status by Complaint Type

top5 = df["Complaint Type"].value_counts().head(5).index

plot_df = df[df["Complaint Type"].isin(top5)]

plt.figure(figsize=(12,6))

sns.countplot(
    data=plot_df,
    y="Complaint Type",
    hue="Status"
)

plt.title("Complaint Status by Complaint Type")
plt.show()

print("="*60)
# Average Resolution Days by Complaint Type

resolution = (
    df.groupby("Complaint Type")["Resolution Days"]
      .mean()
      .sort_values(ascending=False)
      .head(10)
      .reset_index()
)

plt.figure(figsize=(12,6))

sns.barplot(
    data=resolution,
    x="Resolution Days",
    y="Complaint Type"
)

plt.title("Average Resolution Days by Complaint Type")
plt.show()

print("="*60)
# Average Resolution Days by Borough

borough_resolution = (
    df.groupby("Borough")["Resolution Days"]
      .mean()
      .sort_values(ascending=False)
      .reset_index()
)

plt.figure(figsize=(10,5))

sns.barplot(
    data=borough_resolution,
    x="Resolution Days",
    y="Borough"
)

plt.title("Average Resolution Days by Borough")
plt.show()
```
